```python
import cv2
import numpy as np
import imutils
import matplotlib.pyplot as plt

# Load the original image
image_path = "image_with_rectangles.jpg"  # Replace with the actual image path
# Read the image
image = cv2.imread(image_path)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')
plt.show()

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()

# Apply thresholding to create a binary image
# Threshold value of 200 us used to create a binary image with objects in white and background in black
_, binary_image = cv2.threshold(gray_image, 200, 255, cv2.THRESH_BINARY_INV)
# Define a kernel for erosions
kernel = np.ones((5, 5), np.uint8)

# Using cv2.erode() method
# erosion removes pixels on object boundaries
binary_image = cv2.erode(binary_image, kernel)
plt.imshow(binary_image, cmap='gray')
plt.title("Binary Image")
plt.axis('off')
plt.show()

# Find contours in binary image
# Contours are simply the boundaries of white regions in the image
# RETR_EXTERNAL retrieves only the extreme outer contours, discarding nested ones
[contours,hierarchy] = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# Create a canvas to draw contours on
# A canvas is a blank image on which we'll draw the identified contours
contour_image = np.full((image.shape[0], image.shape[1],3), 255, np.uint8)
# Draw contours on the canvas
# This adds green lines around the identified objects in the image
cv2.drawContours(contour_image, contours, -1, (0,255,0), 3)

# Display the image with drawn contours using matplotplib 
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title("Contour Image")
plt.axis('off')
plt.show()
```

```python
# Create a canvas to place aligned rectangles on
blank_image = np.full((image.shape[0], image.shape[1],3), 255, np.uint8)

# Function to extract rotated sub-image from the original image
def getSubImage(rect, src):
    # Get center, size, and angle from rectangle
    center, size, theta = rect 
    center, size = tuple(map(int, center)), tuple(map(int, size))
    size = tuple(map(lambda x: x + 100, size))
    # Get rotation matrix for rectangle
    M = cv2.getRotationMatrix2D(center=center, angle=theta, scale=1) 
    # Perform rotation on original image
    dst = imutils.rotate(src, theta, center=center)
    # Extract the rotated sub-image
    out = cv2.getRectSubPix(dst, size, center)   
    return out

# Iterate through each contour and process rotated rectangles
for contour in contours:
    # Get minimum area that bounds the contour
    rect = cv2.minAreaRect(contour)
    out = getSubImage(rect, image)
    center = rect[0]
    size = rect[1]
    x_offset = int(center[0] - max(size) / 2)
    y_offset = int(center[1] - min(size) / 2)
    theta = rect[2]
    
    # If the angle is greater than 45 degrees, rotate the output image
    if theta > 45:
        out = cv2.rotate(out, cv2.ROTATE_90_CLOCKWISE)
    # Place the rotated sub-image on the canvas
    blank_image[y_offset:y_offset+out.shape[0], x_offset:x_offset+out.shape[1]] = out
        
# Display the final image with aligned rectangles
plt.imshow(cv2.cvtColor(blank_image, cv2.COLOR_BGR2RGB))
plt.title("Aligned Rectangles")
plt.axis('off')
plt.show()
```
